```python
import ROOT
import particle as pt
import funx as fx
import numpy as np
```

```python
H0 = ROOT.TF1("null hypothesys poisson" , "TMath::Poisson(x,2.83)",0, 15)           #Solo GB
H1 = ROOT.TF1("Alternative hypothesys poisson" , "TMath::Poisson(x,5.24)",0, 15)    #BG+Signal
```

```python
c=ROOT.TCanvas()
c.Draw()
H0.SetLineColor(ROOT.kBlue)
H0.Draw()
H1.Draw("SAME")
```

```python
diff = ROOT.TF1("diff", "TMath::Abs(TMath::Poisson(x,2.83) - TMath::Poisson(x,5.24))",0,10)
x_min = diff.GetMinimumX()
x_min
```

```python
alpha = H0.Integral(x_min,15) 
beta = H1.Integral(0 , x_min)

print("alpha = ", alpha)  #Probabilità di rigettare hp solo fondo anche se vera
print("beta = ", beta)    #Probabilità di rigettare hp segnale+fondo anche se vera

print("1-alpha = ", 1-alpha)  
print("1-beta = ", 1-beta)    
```

Possiamo escludere quindi l'ipotesi di solo fondo sapendo di sbagliare al $24.4\%$. Ciò significa che possiamo escludere l'ipotesi di fondo con una accuratezza del $75.6\%$ 

---
L'ipotesi di segnale+fondo è compatibile con il segnale che abbiamo osservato, in quanto 
$$ n_{data} = 8.00\pm 2.83  \qquad n_{signal} = 5.24\pm 2.29 $$


```python
F1 = ROOT.TF1("Data" , "TMath::Poisson(x,8)",0, 15)          
F2 = ROOT.TF1("Signal " , "TMath::Poisson(x,5.24)",0, 15)    

c=ROOT.TCanvas()
c.Draw()
F1.SetLineColor(ROOT.kBlue)
H0.SetLineColor(ROOT.kGreen)
H0.Draw()
F2.Draw("SAME")
F1.Draw("SAME")
```
